# Logistic Regression Parameters

Logistic Regression is a **Supervised Machine Learning Classification Algorithm** used to classify data into **two or more classes**.

It internally uses **Gradient Descent** to optimize the feature weights and minimize the **Log Loss (Cost Function)**.

---

# Creating the Model

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
```

---

# Important Parameters

## 1. penalty

The **penalty** parameter specifies the type of **Regularization** used to reduce overfitting.

By default, Logistic Regression uses **L2 (Ridge) Regularization**.

### Syntax

```python
model = LogisticRegression(
    penalty="l2"
)
```

### Available Options

| Penalty | Regularization | Purpose |
|----------|---------------|---------|
| `"l1"` | Lasso | Performs feature selection by making some weights exactly zero |
| `"l2"` | Ridge (Default) | Shrinks weights and reduces overfitting |
| `"elasticnet"` | L1 + L2 | Combines feature selection and weight shrinking |
| `None` | No Regularization | Model is trained without any penalty |

---

### Example

```python
model = LogisticRegression(
    penalty="l1",
    solver="liblinear"
)
```

> **Interview Point:** Every penalty is **not supported by every solver**.

---

# 2. solver

A **Solver** is an optimization algorithm that helps **Gradient Descent** find the best weights by minimizing the **Log Loss**.

Default Solver

```python
solver="lbfgs"
```

---

## Supported Solvers

| Solver | Supported Penalty | Multiclass Support |
|---------|-------------------|--------------------|
| `lbfgs` | L2, None | ✅ Yes |
| `liblinear` | L1, L2 | ❌ No (Binary only) |
| `newton-cg` | L2, None | ✅ Yes |
| `newton-cholesky` | L2, None | ✅ Yes |
| `sag` | L2, None | ✅ Yes |
| `saga` | L1, L2, ElasticNet, None | ✅ Yes |

---

## Multiclass Classification

For **Binary Classification**, Logistic Regression uses the **Sigmoid Function**.

For **Multiclass Classification**, supported solvers internally use the **Softmax Function**.

```text
Binary Classification

Sigmoid Function

↓

Spam / Not Spam
```

```text
Multiclass Classification

Softmax Function

↓

Class 0

Class 1

Class 2

...
```

---

## Important Note

The **SAG** and **SAGA** solvers work best when all numerical features are on a **similar scale**.

Therefore, always apply **feature scaling** before using these solvers.

Example

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

---

## When to Use SAG and SAGA?

- Large datasets
- Many feature columns
- Faster convergence after scaling

---

# 3. max_iter

`max_iter` specifies the **maximum number of iterations** the solver can perform to find the optimal weights.

Default

```python
max_iter = 100
```

Example

```python
model = LogisticRegression(
    max_iter=500
)
```

If the solver cannot converge within the specified iterations, Scikit-learn displays a warning.

---

# 4. tol (Tolerance)

Tolerance determines **when the optimization process should stop**.

Default

```python
tol = 0.0001
```

If the improvement in loss becomes smaller than the tolerance value, Gradient Descent stops updating the weights.

> **Note:** The default value works well for most problems, so it is rarely changed.

---

# 5. class_weight

`class_weight` is used to handle **imbalanced datasets**.

Default

```python
class_weight=None
```

---

## Balanced Dataset

```
Spam      = 500

Not Spam  = 480
```

The classes are nearly equal.

---

## Imbalanced Dataset

```
Spam      = 800

Not Spam  = 200
```

The model may become biased toward the majority class.

---

## Handling Imbalanced Data

### Method 1

```python
model = LogisticRegression(

    class_weight="balanced"

)
```

The algorithm automatically assigns higher importance to the minority class.

---

### Method 2

Use **Resampling Techniques**

- Oversampling
- Undersampling
- SMOTE

---

# 6. C (Regularization Strength)

`C` controls the **strength of regularization**.

Default

```python
C = 1.0
```

`C` is the **inverse of regularization strength**.

### Interpretation

| C Value | Regularization | Model |
|----------|---------------|-------|
| Small C (0.01, 0.1) | Strong | Simpler model (may underfit) |
| Default C (1.0) | Moderate | Balanced model |
| Large C (10, 100) | Weak | Complex model (may overfit) |

---

# Example

```python
model = LogisticRegression(
    C=0.5
)
```

---

# Complete Example

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(

    penalty="l2",

    solver="lbfgs",

    max_iter=300,

    class_weight="balanced",

    C=1.0
)
```

---

# Important Notes

- Logistic Regression is used for **Classification** problems.
- It predicts **categorical class labels**.
- For Binary Classification, it uses the **Sigmoid Function**.
- For Multiclass Classification, supported solvers use the **Softmax Function** internally.
- It minimizes the **Log Loss (Cost Function)** using an optimization algorithm (solver).
- Scaling improves the performance of **SAG** and **SAGA** solvers.
- `class_weight="balanced"` helps when the dataset is imbalanced.
- `max_iter` controls the maximum number of optimization iterations.
- `C` controls the amount of regularization.

---

# When to Use Logistic Regression?

- Binary Classification
- Multiclass Classification
- Linearly separable data
- Probability prediction

Examples

- Email Spam Detection
- Disease Prediction
- Customer Churn Prediction
- Loan Approval
- Sentiment Analysis

---

# When NOT to Use Logistic Regression?

- Continuous numerical prediction.
- Strongly non-linear decision boundaries (without feature engineering).

For continuous target values, use **Linear Regression** instead.
