WHAT IS REGRESSION?
 - IT IS MACHINE LEARNING OR STATISTICAL TECHNIQUE WHICH PREDICTS CONTINUOUS VALUES BY UNDERSTANDING THE RELATION BETWEEN INDEPENDENT AND DEPENDENT COLUMNS.

TYPES OF REGRESSION
- LINEAR REGRESSION
- POLYNOMIAL REGRESSION
- LOGISTIC REGRESSION

HOW TO FIND RELATION BETWEEN TWO COLUMNS?
1. USING CORR() -> TELLS NEGATIVE/POSITIVE CORR AND HOW STRONG/WEAK THE RELATION IS
2. USING SCATTER PLOT -> CANNOT GET THE EXACT CONCLUSION


REGRESSION
1. X, Y -> IDENTIFY INPUT AND OUTPUT COLUMNS
2. TRAIN & TEST -> SPLIT IN TRAIN AND TEST DATA
3. MMODEL CREATION
4. MODEL TRAINING
5. MODEL EVALUATION
6. PREDICTION

LINEAR REGRESSION

DEFINITION:
- Y = MX + C
- IT IS A SUPERVISED LEARNING REGRESSION TECHNIQUE WHICH PREDICTS CONTINUOUS VALUES BY FINDING BEST-FIT LINE(STRAIGHT LINE) TO UNDERSTAND THE RELATION BETWEEN INDEPENDENT AND DEPENDENT COLUMNS

WHEN DO WE USE LINEAR REGRESSION?
- WHEN THE DATA IS LINEAR OR HIGHLY CORRELATED BETWEEN INPUT AND OUTPUT THEN WE USE LINEAR REGRESSION

***

- RELATION BETWEEN INPUT AND OUTPUT COLUMNS IS STRAIGT LINE, THEN IT IS LINEAR DATA
- TO FIND RELATION BETWEEN TWO COLS IS LINEAR OR NOT, USE SCATTER PLOT AND CORRELATION MATRIX


Y = MX + C
- Y -> PREDICTED VALUES
-  M -> SLOPE
-  X -> NEW INPUT
- C -> INTERCEPT:  BASE OR INITIAL VALUE OF Y WHEN X IS 0 

|CUPS OF COFFEE|ENERGY(100%)|
|---|---|
|1|20|
|2|30|
|3|40|

Y => 10, WHEN X = 0    => (0,10)

> CONCLUSION: AS THE NUMBER OF CUPS INCREASES, ENERGY ALSO INCREASES

WHEN THERE ARE 2 INDEPENDENT COLUMNS(X), THEN WHEN BOTH SHOULD BE 0, THEN C = 5
